```python
import matplotlib.pyplot as plt  
import numpy as np
import math
import pandas as pd
```

```python
real_DQN = pd.read_csv("real_DQN_boxed_water_medium_11-30-2019-213120.csv")
alt_DQN_1 = pd.read_csv("DQN_boxed_water_medium_11-30-2019-223450.csv")
alt_DQN_2 = pd.read_csv("DQN_boxed_water_medium_12-01-2019-012028.csv")
q_learning = pd.read_csv("q_learning_boxed_water_medium_11-30-2019-200557.csv")
baseline = pd.read_csv("baseline_boxed_water_medium_11-30-2019-184820.csv")
```

```python
episodes = real_DQN['ep']
```

```python
plt.plot(episodes, real_DQN["mean episode reward"], episodes, alt_DQN_1["mean episode reward"], alt_DQN_2.index, alt_DQN_2["mean episode reward"
])
plt.legend(["real DQN","alt DQN 1", "alt DQN 2"])
plt.show()
```

```python
plt.plot(episodes, real_DQN["mean episode reward"], q_learning.index, q_learning["mean episode reward"], episodes, baseline["mean episode reward"
])
plt.legend(["DQN","Q-learning", "baseline"])
plt.show()
```

```python
plt.plot(q_learning.index, q_learning["mean episode reward"], 'green', q_learning.index, baseline["mean episode reward"
][:76], 'orange')
plt.legend(["Q-learning", "baseline"])
axis_font = {'fontname':'Arial', 'size':'14'}
plt.xlabel("Episodes", axis_font)
plt.ylabel("Mean Reward", axis_font)
plt.show()
```

```python
plt.plot(q_learning.index, q_learning["delta_t"], 'green', q_learning.index, baseline["delta_t"
][:76], 'orange')
plt.legend(["Q-learning", "baseline"])
plt.xlabel("Episodes", axis_font)
plt.ylabel("Steps", axis_font)
plt.show()
```

```python
plt.plot(episodes, real_DQN["% time spent exploring"], 'blue', episodes, np.full(200, 30), 'green', episodes, np.zeros(200), 'orange')
plt.legend(["DQN","Q-learning", "baseline"])
plt.xlabel("Episodes", axis_font)
plt.ylabel("Time (%)", axis_font)
plt.show()
```
